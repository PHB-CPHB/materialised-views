# 2. We are talent scouts looking to win over some of the best new drivers there are. But we don't want them too old! Write a query that finds the winner of all the races, but only if they are younger than 38 years. The query should give return the date, driver surname, driver age, track time in milliseconds, race name and circuit name for all races.

````python
%%sql
SELECT drivers.surname AS driver_surname, drivers.dob AS driver_age, races.name AS race_name, races.date AS race_date, circuits.name AS circuits_name, results.milliseconds AS milliseconds  FROM results 
JOIN drivers ON drivers.driverid = results.driverid AND drivers.dob > '1980-01-01' AND results.position = 1
JOIN races ON races.raceid = results.raceid
JOIN circuits ON circuits.circuitid = races.circuitid
ORDER BY drivers.dob DESC;
````

# 4. Create a materialized view of your query. Using EXPLAIN ANALYZE try to query the view. Write at least 5 lines of text explaining what's going on and why the query execution time changed.

````python
%%sql CREATE MATERIALIZED VIEW race_winners_cache AS
SELECT drivers.surname AS driver_surname, drivers.dob AS driver_age, races.name AS race_name, races.date AS race_date, circuits.name AS circuits_name, results.milliseconds AS milliseconds  FROM results 
JOIN drivers ON drivers.driverid = results.driverid AND drivers.dob > '1980-01-01' AND results.position = 1
JOIN races ON races.raceid = results.raceid
JOIN circuits ON circuits.circuitid = races.circuitid
ORDER BY drivers.dob DESC;
````

````python
%sql EXPLAIN ANALYZE SELECT * FROM race_winners_cache;
````

As we can see from this result then the time to run the query is a lot faster. In this Select statement we are only running one Seq Scan to get everyting from that view. But what we need to remember is that this view only contains the data we gave is at that time so it is not updated. That is also why this is alot faster i believe. If we should update this we would need to run the query agian, which still takes aroung 30ms to exceute. But this is a great solution if the data that we have created the view on is not changed so much.
